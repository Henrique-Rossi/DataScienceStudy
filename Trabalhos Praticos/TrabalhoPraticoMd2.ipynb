# Elaborando metodos para extrair os  registros de locações existentes  para cada ano(número de linhas, número de colunas)?

```
def ano(x):
    return int(x.split("-")[0])
def mes(x):
    return int(x.split("-")[1])
def dia(x):
    return int(x.split("-")[2])
```

```
df['ano'] = df['datetime'].apply(lambda x: ano(x))
```

# Quantos registros existem para o ano de 2011?

```
df[df['ano']==2011]
```

# Quantos registros existem para o ano de 2012?

```
df[df['ano']==2012]
```

# Quantas locações de bicicletas foram efetuadas em 2011?

```
df[df['ano']==2011]['total_count'].sum()
```

# Quantas locações de bicicletas foram efetuadas em 2012?

```
df[df['ano']==2012]['total_count'].sum()
```
